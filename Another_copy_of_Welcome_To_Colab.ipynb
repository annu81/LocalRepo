```
# plt.plot(x, y , color= 'name', linestyle='lien_style', linewidth = 'value', marker = 'marker symbol', label='label name')
#plt.label('text')
#plt.xlim(start, end)

import matplotlib.pyplot as plt
```

```
months = [1, 2, 3, 4, 5, 6]
sales = [100, 200, 150, 250, 80, 350]
plt.plot(months, sales, color='green', linestyle='--', linewidth='3', marker = 'o', label='2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('2025 Sales report')
plt.legend(loc='upper left', fontsize = 12)
plt.grid(color='black', linestyle = ':', linewidth = 1)
plt.xlim(1, 4)
plt.ylim(0, 500)
plt.xticks([1, 2, 3, 4, 5, 6], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
plt.show()
```
